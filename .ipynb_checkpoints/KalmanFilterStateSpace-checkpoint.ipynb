## State-Space Models and the Kalman Filter

In this notebook we will start by implementing a Kalman filter for the local level time series model. Then we illustrate estimating model parameters via maximum likelihood  using the Klaman filer itself to estimate the likelihood. Finally we verify the results against the statsmodels implementation

### 1. Local Trend Model

The Local trend model for a univariate time series can be described mathematically as

$$
\begin{aligned}
 y_t &= \mu_t + e_t, \quad e_t \sim N(0,\sigma_{e}^2) \\ 
\mu_{t+1} &= \mu_{t} + \eta_{t}, \quad \eta_t \sim N(0,\sigma_{\eta}^2)$
\end{aligned}
$$

where $\mu_t$ is the unobserved trend of the series, $y_t$ is the observed data with measurement noise $e_t$ and $\eta_t$ is the state noise commonly referred to as the innovation. Both noise terms are gaussian and independent of $e_t$

### 2. Statistical inference

In statistical inference we are concerned with the properties of the state $\mu_t$ given the observations where $y_T$ refers to the all the observations from $\left\{ y_t |t = 1, · · · , T  \right\}$. We will not discuss statistical inference in detail but nevertheless it is useful to point out the three types of inference here.

1. **Filtering** : where we are concerned with recovering $\mu_t$ from $F_t$ which is math notation to denote all the information available until time $t$
2. **Prediction**: concerned with forecasting $y_{t+h}$ for $h>0$ given $F_t$
3. **Smoothing**: concerned with estimating $\mu_t$ given $F_T$, where $F_T$ denotes all the information available till time T


An excellent introduction to Kalman filters in the context of time-series models is [Prof. Tsay's lecture notes](http://faculty.chicagobooth.edu/ruey.tsay/teaching/mts/sp2011/lec10-11.pdf). Throughout this notebook we will use notation consistent with this lecture as reference.

We define the following:
* Let $\mu_t$ be the conditional expectation of the state $\mu_t$ given $F_j$ i.e. $E(\mu_t | F_j)$ and let $\Sigma_t$ denote the conditional variance of $\mu_t$
* Let $y_{t|j}$ be the conditional mean of $y_t$ given $F_j$
* Let $\nu_t=y_t-y_{t|t-1}$ be the one step ahead forecast error and let $V_t = Var(\nu_t|F_j)$  Since $\nu_t$ is independent of $F_{t-1}$ its variance conditioned on $F_{t-1}$ is the same as its unconditional variance.

With these notations in place we can derive the Kalman Filter. However we will not be doing so in this notebook and instead just list the results for the Local Trend model and follow it up with an example. 

### 3. Kalman Filter for the Local Trend Model

In addition to the above we need to specify the initial conditions on the state $\mu_1$. We will assume that $mu_1 \sim N(\mu_{1|0},\Sigma_{1|0})$  

$$
\begin{aligned}
 \nu_t &= y_t-y_{t|t-1} \\
 V_t &= \Sigma_{t|t-1} + \sigma_e^2 \\
 K_t &= \Sigma_{t|t-1}/V_t \\
 \mu_{t+1|t}&=\mu_{t|t-1} + K_t \nu_t\\
     \Sigma_{t+1|t} &= \Sigma_{t|t-1} (1-K_t) + \sigma_{\eta}^2, t=1,2,3 \dots T
\end{aligned}
$$

Note: $K_t$ is also commonly known as the Kalman gain.
#### Estimation of model parameters

To begin with we will magically specify model parameters $\sigma_e, \sigma_{\eta}$ and later we will estimate these through maximum likelihood.

###  4. Python implementation

#### Data

As in the lecture referred to above, we use the the  intra-daily  realized  volatility  of  Alcoa  stock  from  January  2,  2003  to  May  7,  2004  for  340 observations.   The  daily  realized  volatility  used  is  the  sum  of  squares  of  intraday  10-minute  log returns measured in percentage.  No overnight returns or the first 10-minute intraday returns are used.  The series used in the demonstration is the logarithm of the daily realized volatility. The data can be found [here](http://faculty.chicagobooth.edu/ruey.tsay/teaching/fts2/aa-3rv.txt)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def loadAlcoaData():
    df=pd.read_csv('aa-3rv.csv')
    df.rename(columns={'10m': 'y_t'},inplace=True)
    df=df[['y_t']]
    df=np.log(df)
    
    return(df)

df=loadAlcoaData()
print(df.head())
```

```python
#create new columns for the Kalman filter
def KalmanFilter_Init(df,mean0,sigma0):
    df['nu_t']=0 #one step forecast error
    df['V_t']=0  #variance of one step forecast error
    df['K_t']=0  #Kalman gain
    df['mu_t']=0 # unobserved state given t-1
    df['Sigma_t']=0 #state variance given t-1
    print(df.head())

    #initialization
    df.loc[0,'mu_t'] = mean0
    df.loc[0,'Sigma_t'] = sigma0
    return(df)

#specify param values
sigmae2=0.4803**2
sigmaeta2=0.0735**2
    
#initialize the filter
df=KalmanFilter_Init(df,0,100)
```

```python
def KalmanFilter(df):
    nrows=len(df.index)
    for t,val in df.iterrows():
        df.loc[t,'nu_t'] = df.loc[t,'y_t'] - df.loc[t,'mu_t']
        df.loc[t,'V_t']=df.loc[t,'Sigma_t'] + sigmae2
        df.loc[t,'K_t']= df.loc[t,'Sigma_t'] / df.loc[t,'V_t']

        if t<nrows-1:
            df.loc[t+1,'mu_t']= df.loc[t,'mu_t'] + df.loc[t,'K_t'] * df.loc[t,'nu_t']
            df.loc[t+1,'Sigma_t']= df.loc[t,'Sigma_t'] *(1-df.loc[t,'K_t']) + sigmaeta2

    return(df)

#run the filter
df=KalmanFilter(df)

#plots
%matplotlib inline
plt.rcParams["figure.figsize"] = (17,7)
#df[['mu_t']].plot(subplots=True,grid=True,sharex=True)   
df[['y_t','mu_t']].plot(grid=True,color=['g','r'])   

#save this for comparison
mu_t_thisnotebook=df['mu_t']
```

### 5. Maximum Likelihood Estimation of model parameters

Earlier we specified model parameter values ($\sigma_e, \sigma_{\eta}$). We now illustrate how to estimate these using maximum likelihood estimation. Following our reference, the maximum likelihood under normality is:

$$
\begin{aligned}
             p(y_1,\dots,y_T|\sigma_e, \sigma_{\eta}) &= p(y_1|\sigma_e, \sigma_{\eta})  \prod_{t=2}^{T}  (y_t|F_{t-1},\sigma_e, \sigma_{\eta}) 
\end{aligned}
$$

Since the later $y_t$ are just linear combinations of normal random variables (the innovations and the initial values), it follows that (see the reference stated above) the log-likelihood function $ln(L(\sigma_e, \sigma_{\eta}))$ evaluates to:

$$
\begin{aligned}
            ln(L(\sigma_e, \sigma_{\eta})) &= -\frac{T}{2}ln(2\pi) - \frac{1}{2}
\sum_{t=1}^{T} \big[ ln(V_t) + \frac{\nu_t^2}{V_t} \big] \end{aligned}
$$


We can now implement this in python and specify the model parameters to be the estimates which maximize the log likelihood function. The good thing about this is that the Kalman state estimates would also be available along with the filter, the bad is that the full filter (or at least the window that is chosen for estimation) would have to be run during each iteration of the maximization  

```python
from numba import jit
import scipy.optimize as optim


def MLE_KFParams():
    
    
    df=loadAlcoaData()
    df=KalmanFilter_Init(df,0,100)
    itercount = {'counter',0}
    
    @jit
    def KalmanFilterwithLL(params,context):
        df= context['df']
        
        #params contains sigmae,sigmaeta,iteration
        sigmae2= params[0]**2 #sigmae**2
        sigmaeta2 = params[1]**2 #sigmaeta**2
        context['itercount'] +=  1

        nrows=len(df.index)
        LL = (-nrows/2) * np.log(2*np.pi)
        for t,val in df.iterrows():
            df.loc[t,'nu_t'] = df.loc[t,'y_t'] - df.loc[t,'mu_t']
            df.loc[t,'V_t']=df.loc[t,'Sigma_t'] + sigmae2
            df.loc[t,'K_t']= df.loc[t,'Sigma_t'] / df.loc[t,'V_t']

            LL += -0.5 * ( np.log(df.loc[t,'V_t']) + df.loc[t,'nu_t']**2/df.loc[t,'V_t'] ) 

            if t<nrows-1:
                df.loc[t+1,'mu_t']= df.loc[t,'mu_t'] + df.loc[t,'K_t'] * df.loc[t,'nu_t']
                df.loc[t+1,'Sigma_t']= df.loc[t,'Sigma_t'] *(1-df.loc[t,'K_t']) + sigmaeta2

        #negative because we are using scipy.optimize.minimize
        #print('iteration: ',context['itercount'],' -LL:', -LL)
        return(-LL)
    
    context={'df':df,'itercount':0}
    options={'maxiter':100,'disp':True}
    itercount=0
    params =[0.2,0.02]
    optimRes = optim.minimize(KalmanFilterwithLL,params,args=(context,), method='Nelder-Mead',options=options)
    #params = optimRes.x
    
    return((df,optimRes))


df,optimRes=MLE_KFParams()
```

#### Note: Model parameters estimated via ML are very close to the initial values we specified

```python
#difference of estimated and initial model parameters
optimRes.x**2 - [sigmae2,sigmaeta2]
```

### 6. Testing
We now compare this notebook's implementation of the local level model by comparing the results with the well-known statsmodels module in python. Note the purpose of this notebook is to demonstrate the Kalman filter algorithm. The use of loops is not optimal and hence this is much slower than the optimized statsmodel implementation.

```python
import statsmodels.api as sm
alcoa_model = sm.tsa.UnobservedComponents(df[['y_t']], 'local level')
modelresults= alcoa_model.fit()
```

```python
df['statsmodels'] = modelresults.predicted_state[0,0:mu_t_thisnotebook.shape[0]]
df['thisnotebook']=df['mu_t']
#modelresults.plot_components(figsize=(13, 6))
df[['statsmodels','thisnotebook']].plot(grid=True)
```

#### Below we check that model parameters estimated via ML are very close to the statsmodels results

```python
print(modelresults.params-optimRes.x**2)
```
